### Работа с изображениями с помощью skimage

```python
from skimage.io import imread, imshow, imsave
from skimage import img_as_float
import skimage
import numpy as np
```

### Загрузка и вывод изображения

```python
img = imread("flowers.jpg")
imshow(img)
```

### Размер изображения

```python
img.shape
```

```python
img.shape[1]
```

### Работа с пикселями и копирование изображения

```python
img[100, 120]
```

```python
img[100, 120, 1]
```

```python
img_copy = img.copy()
img_copy[0:100, 150:250] = [0,0,255]
imshow(img_copy)
```

```python
flower_center = img[80:150, 60:130]
imshow(flower_center)
```

### Обработка изображений. Арифметические операции. Каналы. Вычисление яркости изображения.

```python
img.dtype
```

```python
img_f = img_as_float(img)
```

```python
img.min(), img.max()
```

```python
img_f.min(), img_f.max()
```

```python
imshow(img_f / 2)
```

```python
imshow(img_f / 4)
```

```python
imshow(img_f / 4 + 0.25)
```

```python
imshow(np.clip(img_f * 1.5, 0, 1))
```

```python
imshow(np.clip(img_f * 2 - 0.5, 0, 1))
```

### Работа с каналами изображения

```python
r = img_f[:,:,0]
imshow(r)
```

```python
g = img_f[:,:,1]
imshow(g)
```

```python
b = img_f[:,:,2]
imshow(b)
```

### Объединение каналов

```python
img_rgb = np.dstack((r, g, b))
imshow(img_rgb)
```

```python
mod_img = np.dstack((b, g, b))
imshow(mod_img)
```

```python
mod_img = np.dstack((r, b, g))
imshow(mod_img)
```

### Усреднение цветов изображения

```python
avg_gray = (r + g + b) / 3
imshow(avg_gray)
```

### Превращение изображения в серый цвет - стандартные коэффициенты: 
\begin{equation}
0.2126*r+0.7152*g+0.0722*b
\end{equation}

```python
gray_img = 0.2126 * r + 0.7152 * g + 0.0722 * b
imshow(gray_img)
```

### Использование встроенной функции

```python
imshow(skimage.color.rgb2gray(img))
```

### Работа с изображениями в OpenCV

```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
image = cv.imread("colors.jpg")
rgb_image = cv.cvtColor(image, cv.COLOR_BGR2RGB)
plt.imshow(rgb_image)
```

```python
plt.imshow(rgb_image[55:110, 120:170])
```

```python
rgb_copy = rgb_image.copy()
rgb_copy[55:110, 120:170] = [0, 255, 0]
plt.imshow(rgb_copy)
```

```python
r, g, b = cv.split(rgb_image)
plt.imshow(r)
```

```python
plt.imshow(g)
```

```python
plt.imshow(b)
```

### Объединение каналов

```python
plt.imshow(cv.merge([r, g, b]))
```

```python
plt.imshow(cv.merge([b, r, g]))
```

```python
plt.imshow(cv.merge([r, b, g]))
```

```python
plt.imshow(cv.cvtColor(rgb_image, cv.COLOR_RGB2GRAY), cmap='gray')
```
